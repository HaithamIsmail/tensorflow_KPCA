```python
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt
from kpca import KPCA
```

```python
# Plot KPCA embedding for 3 dimensions
def plot_kpca(x1,x2):
    ax = plt.axes(projection='3d')
    markersize_1 = 12
    markersize_2 = 3
    
    # X1
    ax.scatter(x1[:, 0], x1[:, 1], x1[:, 2], c='g', marker='s',  s=markersize_1)
    
    # X2
    ax.scatter(x2[:, 0], x2[:, 1], x2[:, 2], c='c', marker='o',  s=markersize_2)
    
    plt.show()
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, random_state=123)

plt.scatter(X[y==0, 0], X[y==0, 1], color='red', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', alpha=0.5)

plt.title('A nonlinear 2Ddataset')
plt.ylabel('y coordinate')
plt.xlabel('x coordinate')

plt.show()
```

```python
scikit_kpca = KernelPCA(n_components=3, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)

plt.scatter(X_skernpca[y==0, 0], X_skernpca[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], X_skernpca[y==1, 1], color='blue', alpha=0.5)

plt.text(-0.48, 0.35, 'gamma = 15', fontsize=12)
plt.title('First 2 principal components after RBF Kernel PCA via scikit-learn')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
plot_kpca(X_skernpca[y==0], X_skernpca[y==1])
```

```python
my_kpca = KPCA(n_components=3, gamma=15)
X_mykpca = my_kpca.fit_transform(X)

X_mykpca = X_mykpca.numpy()

plt.scatter(X_mykpca[y==0, 0], X_mykpca[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_mykpca[y==1, 0], X_mykpca[y==1, 1], color='blue', alpha=0.5)

plt.text(-0.48, 0.35, 'gamma = 15', fontsize=12)
plt.title('First 2 principal components after RBF Kernel PCA via tf implementation')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
plot_kpca(X_mykpca[y==0], X_mykpca[y==1])
```
